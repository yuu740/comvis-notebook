## Histogram Equilization


```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('cat.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Cat Image', gray)
cv2.waitKey(0)
```

```python
equ = cv2.equalizeHist(gray)
cv2.imshow('Equalizarion Image', equ)
cv2.waitKey(0)
# Equ merupakan gamnbar yang sudah kita equalize histogram (contrast)
cv2.destroyAllWindows() # => ini untuk menghilangkan error code
```

```python
import numpy as np
```

```python
res = np.hstack((gray, equ))
cv2.imshow('Result', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Plotting

```python
# Mengambil height dan width dari gambar
h = img.shape[0] 
w = img.shape[1]
gray_counter = np.zeros(256, dtype=int)

for i in range(h):
    for j in range(w):
        gray_counter[gray[i][j]] += 1

equ_counter = np.zeros(256, dtype=int)
for i in range(h):
    for j in range(w):
        equ_counter[equ[i][j]] += 1

print(img.shape) # lebar, tinggi, rgb
```

```python
plt.figure(1, figsize=(8,8))

plt.subplot(2, 1, 1) # row, columns, first plot
plt.plot(gray_counter, 'r', label='Before')
plt.legend(loc = "upper left")
plt.ylabel('qty')
plt.xlabel('intensity')
plt.axis((0, 256, 0, 5000))

plt.subplot(2, 1, 2)
plt.plot(gray_counter, 'b', label='After')
plt.legend(loc= 'upper left')
plt.ylabel('qty')
plt.xlabel('intensity')
plt.axis((0, 256, 0, 5000))
plt.show()
```

```python
clahe = cv2.createCLAHE()
clahe_image = clahe.apply(gray)
cv2.imshow("Clahe Image", clahe_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
